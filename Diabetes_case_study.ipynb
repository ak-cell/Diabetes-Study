Now, analysing the dataset using LOGISTIC REGRESSION model

```python
#Train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Outcome',axis=1),df['Outcome'],test_size=0.30)
```

```python
#Training and Predicting
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
predictions=logmodel.predict(X_test)
```

```python
logscore=logmodel.score(X_test,y_test)
print(logscore)
```

Now analysing the model using KNN

```python
#Importing knn classifier
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
print(knn.score(X_test,y_test))
accuracy_knn=knn.score(X_test,y_test)
```

Now analysing our dataset using SVC

```python
#Support vector classifier
from sklearn.svm import SVC
svclassifier=SVC(kernel='linear')
svclassifier.fit(X_train,y_train)
print(svclassifier.score(X_test,y_test))
pred_svc=svclassifier.score(X_test,y_test)
pred_svc
```

    Now analysing our dataset using Random Forest model

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier
#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)
#Train the model using the training setsy_pred=clf.predict(X_test)
clf.fit(X_train,y_train)
Y_pred=clf.predict(X_test)
print(clf.score(X_test,y_test))
```

```python
Y_pred=clf.predict(X_test)
Y_pred
```

```python
accuracy_rf=clf.score(X_test,y_test)
print('Accuracy of Random Forest',accuracy_rf)
```

Now analysing the dataset using Naive Bayes

```python
#Import Gaussian naive bayes
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,y_train)
print(model.score(X_test,y_test))
naive=model.score(X_test,y_test)
naive
```
